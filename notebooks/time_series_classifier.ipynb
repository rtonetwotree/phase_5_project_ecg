```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report

# Assuming the last column of your DataFrame is the labels
y = fazeli_mitbih_train_df.iloc[:, -1].values  # If using the original non-normalized data for labels

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_fft_features, y, test_size=0.2, random_state=42)

# Define the model pipeline
pipeline = Pipeline([
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])

# Train the model
pipeline.fit(X_train, y_train)

# Make predictions
y_pred = pipeline.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# Peak Detection

```python
import numpy as np
from scipy.signal import find_peaks

def detect_r_peaks(ecg_signal):
    # Adjust 'height' and 'distance' based on your ECG signal characteristics and sampling rate
    peaks, _ = find_peaks(ecg_signal, height=0.2, distance=50)  # Example values, adjust as needed
    return peaks

def calculate_rr_intervals(r_peaks, sampling_rate=125):
    # Calculate RR intervals in seconds, given the sampling rate
    rr_intervals = np.diff(r_peaks) / sampling_rate
    return rr_intervals

# Example application on an ECG signal
sampling_rate = 125  # Given as part of your dataset info

# Assuming 'ecg_signal' is loaded with your ECG data for one sample
r_peaks = detect_r_peaks(ecg_signal)
rr_intervals = calculate_rr_intervals(r_peaks, sampling_rate)

print("RR intervals (s):", rr_intervals)
```

```python
import matplotlib.pyplot as plt

# Try adjusting these values based on your signal characteristics
adjusted_height = 0.2  # Example adjustment
adjusted_distance = int(0.3 * sampling_rate)  # For 200 bpm

# Detect peaks with adjusted parameters
adjusted_peaks, _ = find_peaks(ecg_signal, height=adjusted_height, distance=adjusted_distance)

# Plot to verify peak detection
plt.plot(ecg_signal, label="ECG Signal")
plt.plot(adjusted_peaks, ecg_signal[adjusted_peaks], "x", label="Detected Peaks")
plt.legend()
plt.title("ECG Signal with Adjusted Peak Detection")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Function to detect R-peaks and plot an ECG signal with detected R-peaks
def plot_ecg_with_r_peaks(ecg_signal, sampling_rate, adjusted_height=0.2, adjusted_distance=75):
    # Detect peaks with adjusted parameters
    peaks, _ = find_peaks(ecg_signal, height=adjusted_height, distance=adjusted_distance)

    # Plot ECG signal
    plt.figure(figsize=(12, 4))
    plt.plot(ecg_signal, label="ECG Signal")
    
    # Overlay R-peaks
    plt.plot(peaks, ecg_signal[peaks], "x", color='red', label="Detected R-peaks")

    plt.title("ECG Signal with Detected R-peaks")
    plt.xlabel("Samples")
    plt.ylabel("Amplitude")
    plt.legend()
    plt.show()

    # Calculate and print RR intervals for inspection
    rr_intervals = calculate_rr_intervals(peaks, sampling_rate)
    print("RR intervals (s):", rr_intervals)

# Example: Plotting and inspecting the first few ECG signals
num_signals_to_plot = 3  # You can adjust this number
sampling_rate = 250  # Your ECG signal's sampling rate

for i in range(num_signals_to_plot):
    ecg_signal = fazeli_mitbih_train_df.iloc[i, :-1].values  # Select the ECG signal (excluding the label)
    plot_ecg_with_r_peaks(ecg_signal, sampling_rate)
```
